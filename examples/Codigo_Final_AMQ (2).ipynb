## **Flex Matrix**

$$\Delta = \int_{0}^{L} \frac{M_l \cdot M_u}{EI} \, dx$$

Donde:
- $M_l = 1$
- $M_u = 1$
- $E$ = Módulo de elasticidad
- $I$ = Momento de inercia

#### Coeficiente de Flexibilidad correspondiente a 1 debido a 1 del elemento 1
($F_{11}[1]$)
7.png
$$F_{11} = \int_{0}^{L} \frac{\left(1 - \frac{x}{L}\right)^2}{EI} \, dx$$

$$F_{11} = \frac{1}{EI} \int_{0}^{L} \left(1 - \frac{x}{L}\right)^2 dx$$

$$F_{11} = \frac{1}{EI} \int_{0}^{L} \left(1 - \frac{2x}{L} + \frac{x^2}{L^2}\right) dx$$

$$F_{11} = \frac{1}{EI} \left[ x - \frac{x^2}{L} + \frac{x^3}{3L^2} \right]_{0}^{L}$$

$$F_{11} = \frac{1}{EI} \left( L - \frac{L^2}{L} + \frac{L^3}{3L^2} \right)$$

$$F_{11} = \frac{1}{EI} \left( L - L + \frac{L}{3} \right)$$

$$F_{11} = \frac{1}{EI} \cdot \frac{L}{3}$$

$$\boxed{F_{11} = \frac{L}{3EI}}$$

#### Coeficiente de Flexibilidad correspondiente a 2 debido a 1 del elemento 1
($F_{21}[1]$)
8.png
$$-F_{21} = \int_{0}^{L} \frac{\left(1 - \frac{x}{L}\right)\left(\frac{x}{L}\right)}{EI} \, dx$$

$$-F_{21} = \frac{1}{EI} \int_{0}^{L} \left(\frac{x}{L} - \frac{x^2}{L^2}\right) dx$$

$$-F_{21} = \frac{1}{EI} \left[ \frac{x^2}{2L} - \frac{x^3}{3L^2} \right]_{0}^{L}$$

$$-F_{21} = \frac{1}{EI} \left( \frac{L^2}{2L} - \frac{L^3}{3L^2} \right)$$

$$-F_{21} = \frac{1}{EI} \left( \frac{L}{2} - \frac{L}{3} \right)$$

$$-F_{21} = \frac{1}{EI} \cdot \frac{L}{6}$$

$$\boxed{-F_{21} = \frac{L}{6EI}}$$

### Entonces, para el Caso 1 cuando $Q_1 = 1$ su matriz es:
$$

f_{21} \\
f_{31}
\end{bmatrix}
=
\begin{bmatrix}

0 + 0 + 0
\end{bmatrix}
\rightarrow
\begin{cases}
f_{11} = \dfrac{L}{3EI} \\[10pt]
f_{21} = -\dfrac{L}{6EI} \\[10pt]
f_{31} = 0
\end{cases}
$$

$$

##### Displacement vector corresponding to the redundants in the real structure $\{D_Q\}$

### Código:

```python
DQ = np.array([[-dR1],
               [0],
               [0]])

DQ_df = pd.DataFrame(DQ, columns=['DQ'], index = ['Q1','Q2','Q3'])
DQ_df
```

##### Displacement vector corresponding to the redundants due to real loads in the released or statically determinate structure

$$
\{D_{QL}\} = \left\{ \begin{matrix}
-\frac{wL^3}{24EI} \\[15pt]
\frac{wL^3}{24EI} \\[15pt]
\frac{wL^3}{16EI}
\end{matrix} \right\}
$$

```python
DQL = np.array([[-w*L**3 / (24*E*I)],
                [w*L**3 / (24*E*I)],
                [w*L**3 / (16*E*I)]])

DQL_df = pd.DataFrame(DQL, columns=['DQL'], index = ['Q1','Q2','Q3'])
DQL_df
```

##### Displacement vector corresponding to the redundants due to temperature effects in the released or statically determinate structure

$$
\{D_{QT}\} = \left( \frac{\alpha(T_1 - T_2)}{h} \right) \left\{ \begin{matrix}
-\frac{L}{2} \\[15pt]
L \\[15pt]
L
\end{matrix} \right\}
$$

```python
DQT = L /2 * (alfa*(T1-T2)) / vh * np.array([[-1],
                                             [2],
                                             [2]])

DQT_df = pd.DataFrame(DQT, columns=['DQT'], index = ['Q1','Q2','Q3'])
DQT_df
```

##### Displacement vector corresponding to the redundants due to initial deformations in the released or statically determinate structure

$$
\{D_{QP}\} = \left\{ \begin{matrix}
0 \\[15pt]
\frac{d_p}{L} \\[15pt]
-\frac{2d_p}{L}
\end{matrix} \right\}
$$

```python
DQP = np.array([[0],
                [dP/L],
                [-2*dP/L]])

DQP_df = pd.DataFrame(DQP, columns=['DQP'], index = ['Q1','Q2','Q3'])
DQP_df
```

##### Displacement vector corresponding to the redundants due to support displacements in the released or statically determinate structure

$$
\{D_{QR}\} = \begin{bmatrix}
-\frac{d_{R2}}{L} & 0 \\[15pt]
-\frac{2d_{R2}}{L} & \frac{d_{R3}}{L} \\[15pt]
\frac{d_{R2}}{L} & -\frac{2d_{R3}}{L}
\end{bmatrix}
$$

```python
DQR = np.array([[-dR2/L],
                [-2*dR2/L + dR3/L],
                [dR2/L - 2*dR3/L]])

DQR_df = pd.DataFrame(DQR, columns=['DQR'], index = ['Q1','Q2','Q3'])
DQR_df
```

#### Total displacement vector corresponding to the redundants in the released or statically determinate structure

$$\{D_{QS}\}=
\{D_{QL}\}
+
\{D_{QT}\}
+
\{D_{QP}\}
+
\{D_{QR}\}
$$

```python
DQS = DQL + DQT + DQP + DQR

DQS_df = pd.DataFrame(DQS, columns=['DQS'], index = ['Q1','Q2','Q3'])
DQS_df
```

#### Calculation of redundants

$$\{Q\}=[F]^{-1}\left(\{D_Q\}-\{D_{QS}\}\right)
$$

```python
#------------Due all deformations---------------
Q_all = np.linalg.inv(F) @ (DQ - DQS)
Q_all_df = pd.DataFrame(Q_all, columns=['Q due all deformations'])
#------------only due real loads---------------
Q_L = np.linalg.inv(F) @ (0 - DQL)
Q_L_df = pd.DataFrame(Q_L, columns=['Q due real loads'])
#------------only due climate effects---------------
Q_T = np.linalg.inv(F) @ (0 - DQT)
Q_T_df = pd.DataFrame(Q_T, columns=['Q due climate effects'])
#------------only due previus defformations---------------
Q_P = np.linalg.inv(F) @ (0 - DQP)
Q_P_df = pd.DataFrame(Q_P, columns=['Q due previus defformations'])
#------------only due support diplacement---------------
Q_R = np.linalg.inv(F) @ (0 - DQR)
Q_R_df = pd.DataFrame(Q_R, columns=['Q due support diplacement'])

Q = pd.concat([Q_all_df, Q_L_df, Q_T_df, Q_P_df, Q_R_df], axis = 1, ignore_index= False)
Q
```

### **Sumatoria de Fuerzas**

$$
\begin{aligned}

\end{aligned}

$$
