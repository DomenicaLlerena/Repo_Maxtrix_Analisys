#### Data

```python
E = 21000000
l = 3
L = [l,l,l*np.sqrt(2),l*np.sqrt(2),l,l]
thethas = [0,90,135,45,90,0]
db = 0.018
A = np.pi * db**(2) / 4
local_dof = 2                                                                                                       # Local DOF per element
Global_dof = 4                                                                                                      # Local DOF per element
```

#### Solution

##### Stiffness Matrix in Local Coordinates

```python
# --- Create structure container -----------------------------------------------------------------------------------
structure = Manager_K_T_elements2D(method= 'ARM')                                                                                       # Create an instance of the MF structure

# --- Add ARM elements to the structure -----------------------------------------------------------------------------
for i in np.arange(0, len(thethas), 1):                                                                                      # Loop to add each element
    structure.add_element(ARM_K_T_Element2D(E = E,A = A, L = L[i], thetha= thethas[i]))   
```

```python
# --- Stack local stiffness matrices --------------------------------------------------------------------------------
K_all = structure.stacked_stiffness_matrices()                                                                      # Stacked stiffness matrix (2N x 2)

# --- Convert to DataFrame for visualization -----------------------------------------------------------------------
n_elem = len(structure.elements)
n_dof  = local_dof

rows = pd.MultiIndex.from_tuples(
    [(f'Elem_{e+1}', f'DOF_L_{d+1}') for e in range(n_elem) for d in range(n_dof)],
    names=['Element', 'Local DOF']
)

K_all_df = pd.DataFrame(K_all, index=rows, columns=[f'DOF_L_{i}' for i in np.arange(1, K_all.shape[1] + 1)])        # Convert to DataFrame for better visualization
K_all_df.head(len(K_all_df))                                                                                        # Display stacked matrices 
```

##### Transformation Matrix

```python
# --- Stack Transformation matrices --------------------------------------------------------------------------------
T_all = structure.stacked_transformation_matrices()                                                                 # Stacked transformation matrix (4N x 2)

# --- Convert to DataFrame for visualization -----------------------------------------------------------------------
n_dof  = Global_dof
rows = pd.MultiIndex.from_tuples(
    [(f'Elem_{e+1}', f'DOF_L_G_{d+1}') for e in range(n_elem) for d in range(n_dof)],
    names=['Element', 'Local - Global DOF']
)

T_all_df = pd.DataFrame(T_all, index=rows, columns=['Column 1', 'Column 2'])                                        # Convert to DataFrame for better visualization
T_all_df.head(len(T_all_df))                                                                                        # Display stacked matrices
```

##### Stiffness Matrix in Global Coordinates

```python
Keg_all = []
for i in np.arange(0,n_elem,1):
    Telem = structure.elements[i].transformation_ARM_matrix_2D()
    Kelem = structure.elements[i].stiffness_matrix_ARM_AE()
    Keg_all.append(Telem@Kelem@Telem.T)
Keg_all
```

##### Location Vectors (Vectores de Colocación)

```python
# Vectores de Colocación (4x1 por cadaelemento) 
# Cada valor indica el DOF GLOBAL al que corresponde el DOF local 

placement_vectors = [
    np.array([1, 2, 3, 4]),   # Elemento 1
    np.array([0, 0, 1, 2]),   # Elemento 2
    np.array([5, 0, 1, 2]),   # Elemento 3
    np.array([0, 0, 3, 4]),   # Elemento 4
    np.array([5, 0, 3, 4]),   # Elemento 5
    np.array([0, 0, 5, 0]),   # Elemento 6
]

# Visualizar vectores de colocación 
loc_df = pd.concat(
    [pd.DataFrame(loc.reshape(4,1), columns=[f'Elem_{i+1}'], index=[f'd{j+1}' for j in range(4)])
     for i, loc in enumerate(placement_vectors)],
    axis=1
)
print('Vectores de Colocación (4x1 por elemento):')
loc_df
```

##### Matriz de Rigidez Global

```python
total_dof   = 8
active_dofs = 5                                              # DOFs activos: 1 a 5
K_structure = np.zeros((total_dof, total_dof))

for e in range(n_elem):
    Keg       = Keg_all[e]                                   # Matriz 4x4 del elemento e en coord. globales
    placement = placement_vectors[e]                          # Vector de colocación (4,) del elemento e
    for i_local in range(4):
        for j_local in range(4):
            i_global = placement[i_local]
            j_global = placement[j_local]
            if i_global != 0 and j_global != 0 \
               and i_global <= active_dofs and j_global <= active_dofs:   # Solo DOFs 1-5
                K_structure[i_global - 1, j_global - 1] += Keg[i_local, j_local]

# Esto es estetico para Visualización 
dof_labels = [f'D{i+1}' for i in range(total_dof)]
K_structure_df = pd.DataFrame(K_structure, index=dof_labels, columns=dof_labels)
print('Matriz de Rigidez Global K_structure [8x8] (DOFs activos: D1-D5, D6-D8 = 0):')
K_structure_df
```
